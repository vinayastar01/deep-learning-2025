```python
import pandas as pd
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df
```

```python
train_df_shuffled = train_df.sample(frac=1, random_state=42)
train_df_shuffled.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python

train_measures, test_measures, train_labels, test_labels = train_test_split(train_df_shuffled["text"].values,
                                                                            train_df_shuffled["target"].values,
                                                                            test_size=0.1,
                                                                            random_state=42)
```

```python
len(train_measures)
```

Model 0

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
model_0 = Pipeline([
    ("tfidf", TfidfVectorizer()), # convert word to number
    ("clf", MultinomialNB())#model the text
])
model_0.fit(train_measures, train_labels)
```

```python
baseline_score = model_0.score(test_measures, test_labels)
print(f"Our baseline model achieves an accuracy of: {baseline_score*100:.2f}%")
```
